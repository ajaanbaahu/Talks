 $$F : U \times I \to R $$

### Goal of a recommendation system

*Given a partially filled matrix of ratings ($|U|\times|I|$), estimate the missing values.*


$$ 
\newcommand{\argmax}{\mathop{\rm argmax}\nolimits}
\forall{u \in U},\; i^* = \argmax_{i \in -I(u)} [S(u,i)] 
$$


### Challenges

- Gathering known ratings for matrix
    
    - How to collect data in the utility matrix.
        - Explicit 
            - Ask people to rate items. This method doesn't scale.
    
        - Implicit

- Extrapolate unknown ratings from known ones

    - Cold start

## Content Based System

### Goal :

- Recommend items to a user similar to items previously rated by the user

User taste is matched against the items in the catalogue to find the item which matches the user's taste the most.

### Example

- Movies : Same actor(s), director(s), genere etc.
- Websites, blogs, news : Articles with "similar" content
- People : Recommend people with many common friends

### Item Profiles

- For each item, create an item profile e.g. text features as important words in a document
- Profile is a set of features e.g. author, actor, director, metadata, tags etc.
- Convenient to think of the item profile as a vector (boolean or real-valued)

### User Profiles - built by aggregating profiles of items rated 

- User has rated items with $i_1$, $i_2$,..$i_n$
- Could be a simple weighted average of rated item profiles
- More sophisticated aggregations possible



### Example

Assume items are movies, only feature is "Actor"

$Item Profile$ - Description of items (vector of 0 or 1 for each actor)

Suppose user $x$ has watched 5 movies :

- 2 Movies featuring actor A 
- 3 Movies featuring actor B

$User Profile$ = mean of item profiles

- Feature A's weight = 2/5 = 0.4
- Feature B's weight = 3/5 = 0.6


$$ 
\newcommand{\aggr}{\mathop{\rm aggr}\nolimits}
r_{u,i} = \aggr_{i' \in I(u)} [r_{u,i'}] 
$$

### Content-based: generalizations of the aggregation function

The above expression does not take into account the similarity between the item to be recommended and items that a user has rated. Hence incorporating metadata about items, which makes the term 'content' make more sense the following expression is more robust:

$$ sim(u,i) = \frac {w_u . w_i} {||w_u||*||w_i||}$$

$sim$ is a measure of similarity between the item a user has rated and the items to be recommended. Higher the $sim$ measure, more likely will be the item to be recommended to the user

$w_u$ is the user profile vector and $w_i$ is the item profile vector

### Advantages of Content-based approaches

- Able to recommend to users with unique tastes
- Able to recommend new and un-popular items
- Clear explanations for recommended item
- No need for data on other users
- Easy computation

### Disadvantages of Content-based approaches

- Finding appropriate features is hard
- Highly simplified and cannot handle interdependencies
- Never recommends items outside user's content profile 
- Unable to exploit quality judgements of other users
- Cold start problem for new users

## Collaborative filtering


Collaborative Filtering(CF) systems work by collecting user feedback in the form of ratings for items in a given domain and exploiting similarities in rating behaviour amongst several users in determining how to recommend an item.

There are two types of CF systems, neighborhood-based and model based.

### Neighborhood-based CF Systems

### User-Based

In this approach a subset of users are chosen based on their similarity to an active user and a weighted combination of their ratings is used to produce predictions for this user.

### Goal
For a user $u$, recommend item $i$ rated by a set of user $K$ who are in the neighborhood of $u$

### Non-personalized Recommendations

$$p_{a,i}=\frac{\sum_{u=1}^{K}(r_{u,i})}{n} $$


### Personalized Recommendations

Instead of drawing evenly from every one, we would like to sum over all the user ratings for the item times some personalized weight that reflects an agreement between the user and the other users

$$p_{a,i}=\frac{\sum_{i=1}^{I}(r_{u,i})\times sim_{a,u}}{\sum_{u=1}^{K}sim_{a,u}} $$


### What are similar users?

```
|-------------------+-----+-----+-----+-----+-----|
|                   |            Movies           |
|-------------------+-----+-----+-----+-----+-----|
|     Users         | HP1 | HP2 | HP3 | SW1 | SW2 |
|-------------------+-----+-----+-----+-----+-----|
| A                 | 4   | ?   | ?   | 5   | 1   |
|-------------------+-----+-----+-----+-----+-----|
| B                 | 5   | 5   | 4   | ?   | ?   |
|-------------------+-----+-----+-----+-----+-----|
| C                 | ?   | ?   | ?   | 2   | 4   |
|-------------------+-----+-----+-----+-----+-----|
| D                 | ?   | 3   | ?   | ?   | ?   |
|-------------------+-----+-----+-----+-----+-----|
```


- For a user U, find a set N of other users whose ratings are "similar" to U's ratings and estimate U's ratings based on the ratings of users in N.

- From the data we want to find the simlarity metric $sim(A,B)$ between user $A$ with rating vector $R_A$ and user $B$ with rating vector $R_B$
- To capture the intuition that $sim(A,B)$ > $sim(A,C)$


##### Step 1

The weight $w_{a,u}$ which is the measure of similarity between the user $u$ and active user $a$ is calculated using commonly used Pearson correlation coefficient :

$$ w_{a,u}= \frac{\sum_{i=1}^{I}(r_{a,i}-\bar r_a)(r_{u,i}- \bar r_u)}{\sqrt {\sum_{i=1}^{I}(r_{a,i}-\bar r_a)^2}{\sqrt {\sum_{i=1}^{I}(r_{u,i}- \bar r_u)^2}}} $$ 

Here $I$ is the set of items rated by both users, $r_{u,i}$ is the rating given to item $i$ by user $u$ and $\bar r_u$ is the mean rating given by user $u$

Each user is assigned a weight with respect to similarity with the active user

##### Step 2

Ways to choose neighbors:
   
   - Select all neighbors
   - Select top K neighbors
   - Threshold similarity or distance
   - Random selection

$K$ users are selected that have the highest similarity with the active user commonly called the neighborhood

##### Step 3

Predictions $p_{a,i}$ for user $a$ for item $i$ are computed as the weighted average of deviations from the neighbor's mean (i.e. mean centering the data since users can have significant difference between their own rating distributions):

$$p_{a,i}=\bar r_a + \frac{\sum_{u=1}^{K}(r_{u,i}-\bar r_u)\times w_{a,u}}{\sum_{u=1}^{K}w_{a,u}} $$


Instead of using just the deviances from the user's mean we can also use $z$-scores to deal with users ratings spreading across the rating scale

$z$-score for a user's rating for an item :

$$z_{u,i}= \frac {r_{u,i}-\bar {r_u}}{\sigma_u}$$

$$p_{a,i}=\bar r_a + (\frac{\sum_{u=1}^{K}z_{u,i}\times w_{a,u}}{\sum_{u=1}^{K}w_{a,u}})\times \sigma_a $$

### Issues with user-user based approach

- Sparsity
  - Large item sets, small number of ratings, too often there are points where no recommendation can be made since user cannot find neigbors, or the neighbors haven't rated the items
- Computation complexity
  - With milions of users, computing all pairs correlation is expensive
  - User profile changes continuously,hence updating is expensive

### Item-Based

In this approach, rather than matching similar user, a user's rated item is matched to similar items.

- Average item have many more ratings than an average user
- Ratings of an item generally do not change rapidly
- This approach is faster.


##### Step 1

$w_{i,j}$ defines the similarity between pairs of items $i$ and $j$ are calculated offline :

$$ w_{i,j}= \frac{\sum_{u \in U}(r_{u,i}-\bar r_i)(r_{u,j}- \bar r_j)}{\sqrt {\sum_{u \in U}(r_{u,i}-\bar r_i)^2}{\sqrt{ \sum_{u \in U}(r_{u,j}- \bar r_j)^2}}} $$ 


##### Step 2 

A neighborhood set $K$ of the $k$ items which are most similar to the item to be predicted is generated 

##### Step 3

Rating for item $i$ for user $a$ can be predicted using a simple weighted average as:

$$p_{a,i}=\frac{\sum_{j \in K}r_{a,j}w_{i,j}}{\sum_{j \in K}|w_{i,j}|} $$

### Disadvantages of CF

- The principal disadvantage is that CF systems cannot produce recommendations if there are no ratings available

- They demonstrate poor accuracy when there is little data about users’ ratings

## Latent Factor Models

Ratings are deeply influenced by a set of factors that are very specific to the domain such as comedy versus drama, amount of action, or orientation to children etc. These factors are in general not obvious.
An alternative approach that tries to explain the ratings by characterizing both items and users with more holistic goal to uncover the latent features. 

### Matrix Factorization

Just as its name suggests, matrix factorization is to, obviously, factorize a matrix, i.e. to find out two (or more) matrices such that when you multiply them you will get back the original matrix. 

Such a model is closely related to Singular Value Decomposition. 
#### Formalization

- We have a set $U$ of users, and a set $D$ of items. 
- Let $R$ of size $|U| \times |D|$ be the matrix that contains all the ratings that the users have assigned to the items. 

Also, we assume that we would like to discover $K$ latent features. 

- Our task, then, is to find two matrics matrices $P$ (a $|U|$ \* $K$ matrix) and $Q$ (a $|D|$ \* $K$ matrix) such that their product approximates $R$:

$$
\newcommand{\aggr}{\mathop{\rm aggr}\nolimits}
R \approx P \times Q^T = \hat R
$$


- Each row of $P$ would represent the strength of the associations between a user and the features. 
- Each row of $Q$ would represent the strength of the associations between an item and the features.

In the current context matrix factorization decomposes the ratings matrix and maps users and items into a low-dimentsional vector space. Predicted ratings are given by inner products in this space, so for user *u* and item *i* :

$$ 
\newcommand{\aggr}{\mathop{\rm aggr}\nolimits}
\hat r_{u,i} = q_i^Tp_{u} 
$$

#### Difficulties

- High portion of missing values caused by sparsness in the user-item rating matrix
- Conventional SVD is undefined when the knowledge about the matrix is incomplete
- Addressing only the relatively few known entries is highly prone to overfitting

Filling missing ratings and making the matrix dense(e.g. using average ratings for user and item) can be very expensive as :

- It significantly increases the amount of data
- Inaccurate imputation might distorts the data

## Modeling directly the observed ratings only
A solution to the mentioned issues is to model directly the observed ratings :

- Avoids overfitting through a regularized model
- To learn the factor vectors ($p_u$ and $q_i$)the system minimizes the regularized squared error on the set of known ratings:


   $$ \min_{q^*,p^*} \sum_{(u,i)\in K} (r_{ui}-q_i^Tp_u)^2 + \lambda({||q_i||^2} + ||p_u||^2)$$

- Here K is the set of $(u,i)$ pairs for which $r_{u,i}$ is known
- The goal is to generalize those previous ratings in a way that predicts future unknown ratings
- The constant $ \lambda $ controls the extent of regularization

### Adding Bias

- CF data exhibits large systematic tendencies for some users to give higher ratings than others as well as some items to recieve higher ratings that other items 
- It will be unwise to explain the full rating in this and hence user and item bias nedds to be identified
- The bias accounts for the user and item effects

a first order approximation of the bias is as follows:

$b_{ui} = \mu + b_i + b_u$

$\mu$ denotes the overall average rating
$$ $$
$b_u$ = rating deviation of user $u$=(avg.rating of user $u$-$\mu$)
$$ $$
$b_i$ = rating deviation of item $i$=(avg.rating of item $i$-$\mu$)


## Example

- Suppose, average rating overall movies is 3.7 stars. Movie M1 is better than an average movie, so it tends to be rated 0.5 stars above the average movie.
- John is a critical user, who tends to rate 0.3 stars lower than average

Thus the estimate of M1 by John according to the approximation 
$b_{a,i}$
= 3.7+0.5-0.3=3.9

So for item-item based CF:

$$p_{a,i}=b_{a,i}+\frac{\sum_{j \in K}(r_{a,j}-b_{a,j})w_{i,j}}{\sum_{j \in K}|w_{i,j}|} $$


And for matrix factorization model:

$$\hat r_{u,i}=\mu + b_u + b_i+q_i^Tp_{u}$$


The following shows the observed rating broken down into its four components : $global$, $average$, $item-bias$,$user-bias$, and $user-item-interaction$.

$$ \min_{q^*,p^*,b^*} \sum_{(u,i)\in K} (r_{ui}-\mu -b_u-b_i-q_i^Tp_u)^2 + \lambda({||q_i||^2} + ||p_u||^2+b_u^2+b_i^2)$$



## Standard Optimization

There are two methods to minimize the cost function:

- Stochastics Gradient Descent (SGD)
- Alternating Least Squares (ALS)

### Stochastic Gradient Descent 

For each given training case, the system predicts $r_{u,i}$ and computes the associated prediction error :

$$e_{u,i}=r_{u,i}-q_i^Tp_{u}$$

Then it modifies the parameters by a magnitude proportional to $\gamma$ in the opposite direction of the gradient,yielding: 

$$q_i \leftarrow q_i+ \gamma.(e_{u,i}.p_u-\lambda.q_i)$$

$$p_i \leftarrow p_i+ \gamma.(e_{u,i}.q_u-\lambda.p_i)$$

Pros :

- Simple to implement
- Relatively fast to implement

Cons:

- Inherently sequential

### Alternating Least Squares

Because both $q_i$ and $p_u$ are unknowns the objective function is not convex. Given item features, it is easy to find the user features. Similarly, given user features, it is easy to find the item features. 

Using Alternating Least Squares(ALS) the entire loss function can be minimized at once, but, only twiddling half the parameters. That's because the optimization has an easy algebraic solution - if half the parameters are fixed. Half the parametrs are fixed, the other half is recomputed and the process is repeated. There is no gradient in the optimization step since each optimization problem is convex and doesn't need an approximate approach. But, each problem is not the "real" optimization problem since only half the parameters are fixed.

ALS rotates between fixing the $q_i'$s and the $p_u'$s. Assuming all $p_u'$s are fixed, the system recomputes the $q_i'$s by solving the least squares problem, and vice versa to minimize the equation until convergence : 

$$ \min_{q^*,p^*} \sum_{(u,i)\in K} (r_{ui}-q_i^Tp_u)^2 + \lambda({||q_i||^2} + ||p_u||^2)$$


The matrix P can be fixed as some matrix $\hat P$ such that the minimization problem would be equilvalent to $ R= \hat PQ^T $

$$ Q^T = (\hat P^{T} \hat P)^{-1} \hat P^{T}R $$


Analogously $Q$ can be fixed as $\hat Q$, then 

$$ P=R \hat Q(\hat Q^T \hat Q)^{-1} $$

The learning rule then becomes:

$$ P \leftarrow R  Q( Q^T Q)^{-1} $$

$$ Q^T \leftarrow ( P^{T} P)^{-1} P^{T}R $$


### SVD with Implicit Feedback $\rightarrow$ SVD++ 

This approach makes use of implicit feedback information from the user. 

- A user implicitly tells the system about his or her preferences by choosing to declare their opinion and give a high or a low vote
- The rating matix is reduced into a binary matrix where "1" stands for rated and "0" stands for not rated.
- Data not that big as other implicit sources could be, still including this data imrpoves prediction accuracy


Each user is related with two sets of items:

- $r(u)$, which contains all the items for which ratings by $u$ available
- $N(u)$ which contains all the items $u$ provided an implicit preference

Integrating the implicit feedback leads to the following model:

$$\hat r_{u,i}=\mu +b_u+b_i+q_i^T(p_u+\frac {\sum_{j \in N(u)} y_j}{\sqrt{|N(u)|}})$$
